```python
# Importación de librerías pytz y CountryInfo para generar columnas de Zona horaria y Capital de país
import pytz
from countryinfo import CountryInfo
```

```python
# Se crea un filtro de países sin duplicados
paises_df=df_copia['Nationality'].drop_duplicates(keep= 'first')
```

```python
print(paises_df)
```

```python
# Convertimos el filtro en dataFrame, para poder aplicar la función que buscará los códigos ISO
import pandas as pd
paises_unicos= pd.DataFrame(paises_df)
```

```python
paises_unicos
```

```python
# La librería pytz requiere códigos ISO de cada país para generar la zona horaria
# Se aplica una función que agrega al dataFrame una columna de códigos ISO, con CountryInfo.
def iso(pais):
    try:
        return CountryInfo(pais).info().get('ISO').get('alpha2')
    except:
        return 'No disponible'
    
paises_unicos['ISO']= paises_unicos['Nationality'].apply(iso)
```

```python
paises_unicos
```

```python
# Buscamos los países que no tienen ISO en CountryInfo
paises_unicos[['Nationality', 'ISO']].loc[paises_unicos['ISO']=='No disponible']
```

```python
# Búsqueda de todos los países en CountryInfo
countries= CountryInfo().all()
```

```python
# Creación de una lista de sólo los nombres de países en CountryInfo
paises_countryinfo=[]
for country in countries:
    paises_countryinfo.append(country)
```

```python
print(paises_countryinfo)
```

```python
# Importamos la librería re y se crea una expresión regular para buscar algunas coincidencias en la lista de países de CountryInfo
import re
coincidencia= [i for i in paises_countryinfo if re.search(r'republic', i)]
print(coincidencia)
```

```python
# Con estas coincidencias, rellenamos un diccionario con nombres de países actualizados para agregar a los dataFrames.
paises_actualizados= {'England' : 'United Kingdom', 
                    'Bosnia Herzegovina': 'Bosnia and Herzegovina', 
                    'Korea Republic': 'South Korea', 
                    'Montenegro': 'Montenegro', 
                    'Central African Rep.': 'Central African Republic',  
                    'FYR Macedonia': 'Republic of Macedonia', 
                    'China PR': 'China', 
                    'Guinea Bissau': 'Guinea-Bissau', 
                    'Kosovo': 'Kosovo', 
                    'Gambia': 'The Gambia', 
                    'Congo': 'Republic of the Congo', 
                    'Northern Ireland': 'Ireland',
                    'Curacao': 'Curacao',  
                    'Trinidad & Tobago': 'Trinidad and Tobago', 
                    'São Tomé & Príncipe': 'são tomé and príncipe', 
                    'Korea DPR': 'North Korea', 
                    'St Kitts Nevis': 'Saint Kitts and Nevis', 
                    'Antigua & Barbuda': 'Antigua and Barbuda', 
                    'Palestine': 'Palestine', 
                    'St Lucia': 'Saint Lucia', 
                    'Andorra': 'Andorra' 
                      }
```

```python
# Se reemplazan los nombres de los países no encontrados con los valores del dict de países actualizados
paises_unicos= paises_unicos['Nationality'].replace(paises_actualizados)
```

```python
# Se aplica nuevamente la función para buscar los nombres de países que se han actualizado
def iso(pais):
    try:
        return CountryInfo(pais).info().get('ISO').get('alpha2')
    except:
        return 'No disponible'

paises['ISO']= paises['Nationality'].apply(iso)
```

```python
# Buscamos nuevamente los ISO que no existen en CountryInfo
paises[['Nationality', 'ISO']].loc[paises['ISO']=='No disponible']
```

```python
# Lista de países que no han sido encontrados en CountryInfo, si comparamos con el dataFrame de países no duplicados:
paises_no_encontrados=[]
for pais_unico in paises_unicos:
    if pais_unico.lower() not in paises_countryinfo:
        paises_no_encontrados.append(pais_unico)
```

```python
# Datos de DR Congo y Republic of Ireland sí se encuentran en CountryInfo, por lo tanto se actualiza el resto de los países
country=CountryInfo('dr congo').info().get('capital')
print(country)
```

```python
# Función lambda que añade zona horaria, con la librería pytz
import pytz
paises['Zona horaria']= paises['ISO'].apply(lambda x: pytz.country_timezones.get(x))
```

```python
# Función lambda que añade la capital del país de nacionalidad, con CountryInfo
paises['Capital']= paises['ISO'].apply(lambda x: CountryInfo(x).capital())
```

```python
# El código ISO 'RS' es reconocido por Pytz, pero no por CountryInfo
paises[['Nationality', 'ISO']].loc[paises['ISO']=='RS']
```
